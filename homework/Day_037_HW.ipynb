1.線性回歸模型能夠準確預測非線性關係的資料集嗎？

    不行，因為線性回歸在預測分類時，最明顯的一格缺點就是，當有一個極端值時，線性回歸就會受極端值影響很大而偏轉，導致一些原本能預測正確的資料也都不正確了。而加上sigmoid function就很好的解決這點，他將本來在0~1之間的值映射到更接近0或1(小於0.5更接近0，大於0.5更接近1)

2.回歸模型是否對資料分布有基本假設?

    有，其基本假設為資料沒有共線性，若資料有共線性的話，回歸模型就無法處理。更通俗的說，就是資料要是一條線可以描述的。
